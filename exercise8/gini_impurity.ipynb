```python
import pandas as pd

golfdata = pd.read_csv('golf.csv')
golfdata
```

```python
def gini_impurity(feature,label,dataframe):
    rows_overall = dataframe.shape[0]
    feature_counts = dataframe[feature].value_counts().iteritems()
    gini_impurities = []
    gini_weighted = []
    
    for value, count in feature_counts:
        filter_df = dataframe
        filter_df = filter_df[[feature,label]]
        filter_df = filter_df[filter_df[feature] == value]
        
        #Caluclations for sub-node gini
        count_rows = float(count)
        count_yes = len(filter_df[filter_df[label] == "yes"])
        count_no = count_rows-count_yes
        gini_value = ((count_yes/count_rows)*(count_yes/count_rows))+((count_no/count_rows)*(count_no/count_rows))
        gini_impurity = 1-gini_value
        
        #Calculation for weighted gini
        weighted_percent = (count_rows/rows_overall)
        gini_weighted.append(weighted_percent*gini_value)

        gini_impurities.append([value,gini_impurity])
        
    gini_weighted_total = sum(gini_weighted)
    gini_weighted_impurity = 1-gini_weighted_total
    gini_impurities.insert(0,['weighted',gini_weighted_impurity])
    return gini_impurities
```

```python
print (gini_impurity('Temperature','Golf',golfdata))
print (gini_impurity('Outlook','Golf',golfdata))
print (gini_impurity('Humidity','Golf',golfdata))
print (gini_impurity('Windy','Golf',golfdata))
```
